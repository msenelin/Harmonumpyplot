## Asymmetriskt krypto med öppen nyckel

RSA är ett asymmetriskt system med öppen nyckel, där avsändare och mottagare utväxlar krypteringsnycklar. Man har två sammanhörande nycklar. Det som krypteras med den ena kan bara dekrypteras med den andra. Den ena nyckeln kan publiceras offentligt medan man hemlighåller den andra. Det gör inget om kommunikationen avlyssnas, eftersom det bara är mottagaren som kan dekryptera meddelandet. Eller rättare sagt: utan dekrypteringsnyckeln är det “computationally infeasible”, dvs det skulle ta åratal även för den snabbaste superdator att knäcka koden. 

Även om datorerna blir snabbare och snabbare så blir systemet bara säkrare i och med att man då kan använda ännu längre nycklar.
Grejen med detta är alltså att man inte behöver utväxla någon hemlig krypteringsnyckel utan sändare och mottagare behåller motsvarande dekrypteringsnycklar för sig själva. Det tjusiga med detta är att man även kan använda det som digital ID för att bekräfta vem som sänt ett meddelande så att man vet att ingen fejkat det. Om avsändaren använder sin hemliga nyckel för krypteringen kan det endast dekrypteras med avsändarens offentliga nyckel och på så sätt ser man att det inte kan vara någon annan som skickat meddelandet. Man kan göra två krypteringar: först med mottagarens offentliga och sedan med den egna hemliga. Mottagaren dekrypterar i tur och ordning med sin hemliga och sedan med mottagarens offentliga för att se att meddelandet är autentiskt. Bank-ID använder detta för att verifiera avsändaren.

AES är däremot ett så kallat blockchiffer och som sådant oerhört säkert också. Men det är ett symmetriskt chiffer där man använder samma nyckel för kryptering som för dechiffrering. Det innebär att man måste ha en säker kanal för att utväxla denna. För att vara riktigt säker bör man då helst personligen träffa mottagaren och utväxla nyckeln.

### Moduloräkning

I definitionen av variablerna till vänster så använder vi beteckningen

$ a\equiv b \ (mod \ n) $

när vi menar att $a$ och $b$ bildar samma rest vid division med $n$. Vi utläser detta som "$a$ är kongruent med $b$ modulo $n$". Exempelvis är

$ 4 \equiv 7 \ (mod \ 3) $

eftersom 4 och 7 bildar samma rest vid division med 3. Man kan räkna med kongruenstecken och de aritmetiska operationerna nästan som med vanligt likhetstecken.

### Fermats lilla sats

Om p är ett primtal och a är ett heltal så gäller:

$ a^p \equiv a \ (mod \ p) $

<br>


#### ***Bevis av Fermats lilla sats (med induktion)***

Vi visar satsen med hjälp av induktion. 

<ins>Induktionsbas 

Om $a =1$ så är $1^p\equiv\ 1\ (mod\ p)$ och satsen gäller trivialt.

<br>

<ins> Induktionsantagande

Antag att satsen gäller för $ a\ =\ n.$ 

D.v.s så att $ n^p\equiv\ n\ (mod\ p)$

<br>


<ins> Induktionssteg

Om nu $a\ = n+1$ så behöver vi visa att induktionsantagandet medför att 

${(n+1)}^p\equiv\ n+1\ (mod\ p).$

Enligt Binomialsatsen så är:

$ (n+1)^p=\ n^p+\ \binom{p}{1}n^{p-1}\ +\ \binom{p}{2}n^{p-2}+\ ...\ +\ \binom{p}{p-1}n + 1$

Vi skriver om termerna i summan som:

$n^p+1+\sum_{k=1}^{p-1}{\binom{p}{k}n^{p-k}} $
 
 Enligt induktionsantagandet så gäller att $ n^p\equiv\ n\ (mod\ p)$ och således är förstås $ n^p + 1 \equiv\ n + 1 (mod\ p) $$
 
 Det kvarstår således att visa att:
 $$ \sum_{k=1}^{p-1}{\binom{p}{k}n^{p-k}} \equiv 0  \ (mod\ p).$$
 
Binomialkoefficienterna ges av:

$$ \binom{p}{k} = \frac{p!}{k!(p-k)!},$$

kan vi skriva om som: 

$$\frac{p!}{k!(p-k)!} = \frac{p \cdot (p-1) \cdot (p-2) \cdots (p-k+1) \cdot k!}{k!(p-k)!}. $$

Kvar blir alltså:

$$ \frac{p \cdot (p-1) \cdot (p-2) \cdots (p-k+1)}{(p-k) \cdot (p-k-1) \cdots 2 \cdot  1 } $$

och efter som $p$ är ett primtal så finns ingen möjlighet för någon faktor i nämnare att vara en delare av $p.$ Alltså följer att $p$ är en delare av varje term i summan, d.v.s

$$\sum_{k=1}^{p-1}{\binom{p}{k}n^{p-k}} \equiv 0  \ (mod\ p) \quad \quad  \quad \quad \quad \square $$



<br>

**Kommentar:**

En binomialkoefficient $\binom{p}{k}$ betyder per definition antalet sätt att välja $k$ element ur en $p$-mängd (utan ordning) och detta är förstås alltid ett heltal. Att $p$ är ett primtal är avgörande för att $p$ inte ska ha några gemensamma delare med talen som finns i nämnaren. 

<br>

**Övning:**

Hitta ett  exempel $\binom{n}{k}$ där $n$ inte är ett primtal och har en gemensam delare i nämnaren.

### Eulers $\phi$-funktion

Funktionen $\phi(n)$ beräknar antalet positiva heltal mindre än $n$ som är relativt prima till $n$. 

**Exempel**
* 1, 2 är relativt prima till n = 3 och således är $\phi(3)=2$
* 1, 2, 3, 4, 5, 6 är relativt prima till n = 7 och således är $\phi(7)=6$
* 1, 5 är relativt prima till n = 6 och således är $\phi(6)=2$

För vissa specialfall finns enkla metoder för att beräkna $\phi$. 
Om $p$ är primtal så gäller

$\phi(p)= p-1$

$\phi(p^2)=p(p-1)$

```

 # A simple Python3 program  
# to calculate Euler's  
# Totient Function 
  
# Function to return 
# gcd of a and b 
def gcd(a, b): 
  
    if (a == 0): 
        return b 
    return gcd(b % a, a) 
  
# A simple method to evaluate 
# Euler Totient Function 
def phi(n): 
  
    result = 1
    for i in range(2, n): 
        if (gcd(i, n) == 1): 
            result+=1
    return result 
  
# Driver Code 
for n in range(1, 8): 
    print("phi(",n,") = ",  
           phi(n), sep = "") 
```

```
# Python 3 program to calculate 
# Euler's Totient Function 
# using Euler's product formula 
  
def phi(n) : 
  
    result = n   # Initialize result as n 
       
    # Consider all prime factors 
    # of n and for every prime 
    # factor p, multiply result with (1 - 1 / p) 
    p = 2
    while p * p<= n : 
  
        # Check if p is a prime factor. 
        if n % p == 0 : 
  
            # If yes, then update n and result 
            while n % p == 0 : 
                n = n // p 
            result = result * (1.0 - (1.0 / float(p))) 
        p = p + 1
          
          
    # If n has a prime factor 
    # greater than sqrt(n) 
    # (There can be at-most one 
    # such prime factor) 
    if n > 1 : 
        result = result * (1.0 - (1.0 / float(n))) 
   
    return int(result) 
      
      
# Driver program to test above function 
for n in range(1, 11) : 
    print("phi(", n, ") = ", phi(n)) 
     
  
# This code is contributed 
# by Nikita Tiwari. 
```

## RSA-algoritmen

1. Välj två stora primtal $p$ och $q$, $p \neq q$ och beräkna $n = pq$
2. Beräkna nu $\phi(pq) = (p - 1)(q - 1)$
3. Välj $e$ så att $SGD(e, \phi(pq)) = 1$ och $1 < e < \phi(pq)$
4. Beräkna $d$ så att $de \equiv 1 \ (mod \ \phi(pq))$
5. För att kryptera $c=t^e \ (mod \ n)$ och $t=cd \ (mod \ n ) $.

RSA är intressant ur synpunkten att med lite möda kan matematiken förklaras på gymnasienivå. Nyare metoder som med elliptiska kurvor är mycket knepigare men anses ha samma säkerhet och med fördelen av kortare nyckel. Dessa metoder med offentlig nyckel används ju numera för exempelvis bank-ID och säkra uppkopplingar över Internet.

AES är något helt annat och man måste begripa programmering på bitnivå med skift, exclusive-or och annat, och det är inget system med offentlig nyckel.

Men förutom att med lite möda vara begriplig på gymnasienivå är RSA oerhört intressant i och med att det är en metod med “offentlig nyckel”. Detta revolutionerade på 70-talet tänkandet inom kryptografin och finns beskrivet bl.a, i Simon Singhs “The code book”. Svenska översättningen är tyvärr slut på förlaget. Den allmänna idén framkastades av Diffie och Hellman på 70-talet och utvecklades bl.a av Rivest, Shamir och Adleman. Jag råkade då i Scientific Americans augustinummer 1977 läsa en utmaning som de publicerade. Dels en beskrivning av detta med public key. Jag var tvungen att läsa artikeln ett par gånger innan jag insåg vidden av och nytänkandet i detta. Dels publicerades en kort krypterad text samt den offentliga nyckel som använts. De utlyste ett pris på 100 dollar till den som kunde dechiffrera texten. Själva hade de beräknat att det borde ta 40 biljarder år att med den tidens snabbaste datorer knäcka koden, men de visste ju inte säkert om någon skulle kunna hitta en genväg.

Att knäcka koden var ekvivalent med att hitta två primtal som faktoriserade ett 129-siffrigt tal. Dock blev datorerna snabbare och snabbare och matematiker kom på bättre metoder att faktorisera med s.k. talfältssåll. 1994 hade forskarna Atkins, Graff, Lenstra och Leyland knäckt meddelandet. Via en diskussionsgrupp på internet hade de samlat frivilliga som ställt över 1600 datorer av olika kapacitet till förfogande, Dessa fick jobba i över ett halvår med var sin del av problemet och resultaten samlades in för en slutgiltig samkörning på en superdator. Efter två dagar kom resultatet: THE MAGIC WORDS ARE SQUEAMISH OSSIFRAGE. Koden knäcktes visserligen 40 biljarder år tidigare än beräknat, men det tog å andra sidan 17 år av datorutveckling och matematisk forskning, ett halvår med 1600 mindre datorer och två dagar med en superdator för att knäcka koden. Numera finns naturligtvis superdatorer som knäcker det ursprungliga problemet över dagen, eller på några timmar, men RSA är för den skull inte “knäckt”. Det är bara att använda längre nycklar så är kryptot ändå lika svårlöst. Algoritmen bygger inte på någon hemlighet utan bara på att det är i praktiken beräkningsmässigt ogörligt att dekryptera meddelandena.
